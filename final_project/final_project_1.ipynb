# P Value

```python
df = pd.DataFrame(liver)
liver.target=df[["Result"]]
liver.data=df[["Age of the patient",
          "Gender of the patient",
          "Total Bilirubin","Direct Bilirubin",
          "lkphos Alkaline Phosphotase",
          "gpt Alamine Aminotransferase",
          "Sgot Aspartate Aminotransferase",
          "Total Protiens",
          "LB Albumin","A/G Ratio Albumin and Globulin Ratio"]]
liver.feature_names=["Age of the patient",
          "Gender of the patient",
          "Total Bilirubin","Direct Bilirubin",
          "lkphos Alkaline Phosphotase",
          "gpt Alamine Aminotransferase",
          "Sgot Aspartate Aminotransferase",
          "Total Protiens",
          "LB Albumin","A/G Ratio Albumin and Globulin Ratio"]
```

```python
from sklearn import feature_selection as fs
#方法一
#選取K個最好的特徵值
skb=fs.SelectKBest(fs.f_classif,k=9)
skb.fit(liver.data,liver.target)
#Pvalue<0.05為顯著的特徵值
skb.pvalues_
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig,ax = plt.subplots(figsize=(9,7))
plt.plot(liver.feature_names, skb.pvalues_)
plt.xticks(liver.feature_names,rotation=85,fontsize=14) # 設定x軸label以及垂直顯示
plt.xticks(fontsize=14)
fig.autofmt_xdate(rotation=40)
plt.grid(True)
plt.ylabel('P Value',fontsize=14)
plt.xlabel('features',fontsize=14)
```
